# Import Dependencies

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

# Text File Path 

```python
path = "C:/Users/Utkarsh Gaikwad/OneDrive/Desktop/TextInput.txt"
text = open(path,mode='r').read()
print(text)
```

# Stopwords like it , as , a , the etc.

```python
stopwords = STOPWORDS
stopwords
```

# Create Wordcloud

```python
wc = WordCloud(
    background_color='white',
    stopwords = stopwords,
    height=600,
    width=400)
```

```python
wc.generate(text)
```

# Write Wordcloud to .png File

```python
wc.to_file('C:/Users/Utkarsh Gaikwad/OneDrive/Desktop/WordCloud.png')
```
